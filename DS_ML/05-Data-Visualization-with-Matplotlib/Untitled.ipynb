```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,20)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')
axes2.plot(y,x)
```

```python
fig,axes= plt.subplots(1,2)
#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x^2')
ax.plot(x,x**3,'g',label='x^3')
ax.legend(loc=0)
```

```python
fig.savefig('My_Picture.png')
```

```python
fig = plt.figure(figsize=(12,5),dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='--') ##RGB Hex Code
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
